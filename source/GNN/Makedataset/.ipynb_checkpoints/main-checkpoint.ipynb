```python
def TsSplit(ts, L):
    ts_train = [(ts+l) for l in range(L)]
    ts_test = ts_train[-1]+1
    ts_all = ts_train.copy()
    ts_all.extend([ts_test])
    return ts_train, ts_test, ts_all
```

```python
def PrimaryAttributeID():
    for att_id in ATTRIBUTE_TABLE_ROWS:
        if ATTRIBUTE_TABLE[att_id]['primary'] == True:
            return att_id

def AttributeRange(att_id):
    r0 = ATTRIBUTE_TABLE[att_id]['range0']
    r1 = ATTRIBUTE_TABLE[att_id]['range1']
    return r0, r1+1

def ExistNodeList(ts):
    return np.where(EXIST_TABLE[:, ts]==1)[0]

def KnownNodeAttribute(ts):
    return np.load(InputDir + '/node_attribute' + str(ts) + '.npy')

# タイムステップtsの{ノード:Attributes}を取得
def NodeAttributeDict(ts, is_primary=False, att_id=-1):
    all_node_att = KnownNodeAttribute(ts)
    node_list = ExistNodeList(ts)
    if is_primary:
        r0, r1 = AttributeRange(PrimaryAttributeID())
    elif att_id >= 0:
        r0, r1 = AttributeRange(att_id)
    else:
        r0, r1 = 0, all_node_att.shape[1]
    nodes = {}
    for idx, att in enumerate(all_node_att[node_list, r0:r1]):
        nodes[node_list[idx]] = att
    return nodes

# タイムステップtsの{ノード:クラス}を取得
def NodeClassDict(ts, att_id):
    node_class = {}
    for node, attribute in NodeAttributeDict(ts, att_id=att_id).items():
        node_class[node] = attribute.argmax()
    return node_class

# タイムステップtsの{クラス:ノード集合}を取得
def ClassNodeSetDict(ts, att_id):
    r0, r1 = AttributeRange(att_id)
    class_set = {i:set() for i in range(r1-r0)}
    for node, attribute in NodeAttributeDict(ts, att_id=att_id).items():
        class_set[attribute.argmax()].add(node)
    return class_set

# タイムステップtsの{クラス:ノード数}を取得
def ClassNodeNumDict(ts, att_id):
    class_set = ClassNodeSetDict(ts, att_id)
    class_num = {}
    for class_id, node_set in class_set.items():
        class_num[class_id] = len(node_set)
    return class_num

# タイムステップtsの{ノード:pクラス}を取得
def NodePClassDict(ts):
    node_class = {}
    for class_id in range(len(PCLASS_TS_NODE_SET_DICT)):
        for node in PCLASS_TS_NODE_SET_DICT[class_id][ts]:
            node_class[node] = class_id
    return node_class

# タイムステップtsの{pクラス:ノード集合}を取得
def PClassNodeSetDict(ts):
    class_set = {}
    for class_id in range(len(PCLASS_TS_NODE_SET_DICT)):
        class_set[class_id] = PCLASS_TS_NODE_SET_DICT[class_id][ts]
    return class_set

# タイムステップtsの{pクラス:ノード数}を取得
def PClassNodeNumDict(ts):
    class_num = {}
    for class_id in range(len(PCLASS_TS_NODE_SET_DICT)):
        class_num[class_id] = len(PCLASS_TS_NODE_SET_DICT[class_id][ts])
    return class_num
```

```python
def GetAppearedNodes(ts):
    return set(ExistNodeList(ts)) - set(ExistNodeList(ts-1))

def GetObservedNodes(ts, L):
    U = set()
    for i in range(L):
        U |= set(ExistNodeList(ts-i))
    return U

def GetNodes(ts, L, node_type):
    if node_type=='all':
        node_set = set(ExistNodeList(ts))
    elif node_type=='stay':
        node_set = set(ExistNodeList(ts-1)) & set(ExistNodeList(ts))
    elif node_type=='lost':
        node_set = set(ExistNodeList(ts-1)) - set(ExistNodeList(ts))
    elif node_type=='return':
        node_set = GetAppearedNodes(ts) - (GetAppearedNodes(ts) - GetObservedNodes(ts-1, L))
    elif node_type=='new':
        node_set = GetAppearedNodes(ts) - GetObservedNodes(ts-1, L)
    return node_set
```

```python
def LinearRegression(ts_all, ts_train, node_num_train):
    # Learn
    X = pd.DataFrame(ts_train)
    Y = pd.DataFrame(node_num_train)
    model = linear_model.LinearRegression()
    model.fit(X, Y)
    # predict
    px = np.array(ts_all)[:,np.newaxis] # convert to 2D numpy array due to sklearn specification
    py = model.predict(px)
    py = py.reshape(-1).tolist()
    return py

def PlotHistory(history, x, y1, y2, title, ylabel):
    plt.plot(history[x], history[y1], marker=".", color='#1ea8b5', label=y1)
    plt.plot(history[x], history[y2], marker=".", color='#ce405f', label=y2)
    plt.title(title)
    plt.xlabel(x)
    plt.ylabel(ylabel)
    plt.legend()
    plt.show()

def GetPredictedNodeNum(ts_all, ts_train, node_type, plot_title='title'):
    # get num of nodes
    node_num_all = [float(len(NODE_TYPE_TS_NODE_SET_DICT[node_type][ts])) for ts in ts_all]
    node_num_train = node_num_all[:-1]
    # learning & predict (LinearRegression)
    predicted_node_num_all = LinearRegression(ts_all, ts_train, node_num_train)
    # plot result
    #history = {'ts':ts_all, 'true':node_num_all, 'pred':predicted_node_num_all}
    #PlotHistory(history, 'ts', 'true', 'pred', title=plot_title, ylabel="# of node")
    # take out only test years
    predicted_node_num_test = round(0 if predicted_node_num_all[-1] < 0 else predicted_node_num_all[-1])
    return predicted_node_num_test

def SavePredictedNodeNum(ts_all, ts_train):
    num_lost = GetPredictedNodeNum(ts_all, ts_train, 'lost')
    num_return = GetPredictedNodeNum(ts_all, ts_train, 'return')
    num_new = GetPredictedNodeNum(ts_all, ts_train, 'new')
    np.save(OutputDir + '/predicted_node_num/predicted_node_num_' + str(ts_train[0]), np.array([num_lost, num_return, num_new]))

def SaveLabelNodeNum(ts_test):
    num_lost = len(NODE_TYPE_TS_NODE_SET_DICT['lost'][ts_test])
    num_return = len(NODE_TYPE_TS_NODE_SET_DICT['return'][ts_test])
    num_new = len(NODE_TYPE_TS_NODE_SET_DICT['new'][ts_test])
    np.save(OutputDir + '/label_node_num/label_node_num_' + str(ts_train[0]), np.array([num_lost, num_return, num_new]))
    
def GetPredictedNodeNumGroupByClass(ts_all, ts_train, node_type, all_predicted_num):

    def GetNodeNumListGroupByClass(ts_list, node_type, ts_node_set_groupby_class):
        node_num = []
        for ts in ts_list:
            node_num.append(float(len(NODE_TYPE_TS_NODE_SET_DICT[node_type][ts] & ts_node_set_groupby_class[ts])))
        return node_num
    
    predicted_node_num_test_groupby_class = {}
    for class_id, ts_node_set_groupby_class in PCLASS_TS_NODE_SET_DICT.items():
        # get num of nodes
        node_num_all = GetNodeNumListGroupByClass(ts_all, node_type, ts_node_set_groupby_class)
        node_num_train = node_num_all[:-1]
        # learning & predict (LinearRegression)
        predicted_node_num_all = LinearRegression(ts_all, ts_train, node_num_train)
        # plot result
        #plot_title = ATTRIBUTE_TABLE[PrimaryAttributeID()]['class' + str(class_id)]
        #history = {'ts':ts_all, 'true':node_num_all, 'pred':predicted_node_num_all}
        #PlotHistory(history, 'ts', 'true', 'pred', title=plot_title, ylabel="# of node")
        # take out only test years
        predicted_node_num_test = round(0 if predicted_node_num_all[-1] < 0 else predicted_node_num_all[-1])
        predicted_node_num_test_groupby_class[class_id] = predicted_node_num_test

    predicted_node_ratio_groupby_class = {}
    for class_id, predicted_node_num in predicted_node_num_test_groupby_class.items():
        if sum(predicted_node_num_test_groupby_class.values()) == 0:
            predicted_node_ratio_groupby_class[class_id] = 0 # 未知ノードが出現しないグラフの場合
        else:
            predicted_node_ratio_groupby_class[class_id] = round(all_predicted_num * (predicted_node_num / sum(predicted_node_num_test_groupby_class.values())))
    
    return predicted_node_ratio_groupby_class
```

```python
def AppendNewNodes2IdxTable(ts_list):
    for ts in ts_list:
        # get time step list
        ts_train, ts_test, ts_all = TsSplit(ts, L)
        
        # get # of nodes (new, return, lost)
        predicted_node_num_new = GetPredictedNodeNum(ts_all, ts_train, 'new')
        predicted_node_num_new_groupby_class = GetPredictedNodeNumGroupByClass(ts_all, ts_train, 'new', predicted_node_num_new)
        
        # Append new node to NODE_IDX_TABLE
        for class_id, node_num in predicted_node_num_new_groupby_class.items():
            for idx in range(node_num):
                node_idx = len(NODE_IDX_TABLE)
                NODE_IDX_TABLE['new_' + str(class_id) + '_' + str(ts_test) + '_' + str(idx)] = node_idx

def SaveNodeIdxTable(node_idx_table):
    rows = []
    for node, idx in node_idx_table.items():
        rows.append([node, idx])
    df = pd.DataFrame(rows, columns=['node', 'ID'])
    df.to_csv(OutputDir + '/node_idx.csv', index=False, encoding='utf_8_sig')
```

```python
#def GetPredictedNewNodes(ts_test):
#    new_node_set = {ID for node, ID in NEW_NODE_IDX_TABLE.items() if node.split('_')[2] == str(ts_test)}
#    return new_node_set

def GetPredictedNewNodeClassDic(ts_test):
    # new_nodes {node_id:class_id}
    new_node_class_dic = {ID:int(node.split('_')[1]) for node, ID in NEW_NODE_IDX_TABLE.items() if node.split('_')[2] == str(ts_test)}
    return new_node_class_dic

def InitNewNodeAttribute(ts_train, new_node_class_dic, new_node_list):
    new_node_attribute = np.zeros((len(ts_train), len(new_node_list), attribute_size))
    for t_idx, ts in enumerate(ts_train):
        for n_idx, node in enumerate(new_node_list):
            for att_id in ATTRIBUTE_TABLE_ROWS:
                r0, r1 = AttributeRange(att_id)
                if att_id == PrimaryAttributeID():
                    attribute = np.zeros(r1 - r0)
                    attribute[new_node_class_dic[node]] = 1
                #elif att_id == その他の属性:
                #    attribute = 属性初期化値
                new_node_attribute[t_idx, n_idx, r0:r1] = attribute
    return new_node_attribute

def SaveInputGraphNodeAttributes(ts_train, ts_test, L, new_node_class_dic):
    
    # generate numpy matrix of new node attribute
    new_node_set = set(new_node_class_dic.keys())
    new_node_list = sorted(list(new_node_set))
    new_node_attribute = InitNewNodeAttribute(ts_train, new_node_class_dic, new_node_list)

    # generate numpy matrix of all node attribute (merge new to all)
    all_node_attribute = np.zeros((len(ts_train), all_node_num, attribute_size))
    for idx, ts in enumerate(ts_train):
        all_node_attribute[idx, :known_node_num] = KnownNodeAttribute(ts)
    all_node_attribute[:, new_node_list] = new_node_attribute

    #all_node_attribute_reshape = np.zeros((all_node_num * len(ts_train), attribute_size))
    all_node_attribute_reshape = np.zeros((all_node_num, attribute_size * len(ts_train)))
    for t_idx, ts in enumerate(ts_train):
        all_node_attribute_reshape[:, t_idx*attribute_size : (t_idx+1)*attribute_size] = all_node_attribute[t_idx]
    
    sparse_all_node_attribute = sparse.bsr_matrix(all_node_attribute_reshape)
    
    mmwrite(OutputDir + '/time_series_node_attribute/node_attribute_' + str(ts_train[0]), sparse_all_node_attribute)

    return all_node_attribute

def KnownEdge(ts):
    adjacency = mmread(InputDir + '/adjacency' + str(ts) + '.mtx')
    row = adjacency.row
    col = adjacency.col
    nnz = adjacency.nnz
    for idx in range(nnz):
        yield (row[idx], col[idx])
        
def IsUnknownNode(node):
    return node >= known_node_num

def AddObservedEdge(E, ts_train, L):
    for i in range(L):
        for row, col in KnownEdge(ts_train[-1-i]):
            E[row, (L-(i+1))*num_edge_type*all_node_num + col] = 1
    return E

def AddNewEdge(E, ts_train, ts_test, L, new_node_class_dic):
    new_edge_set = set()
    for new_node, new_node_class in new_node_class_dic.items():
        for num, similar_node in enumerate(PCLASS_TS_NODE_SET_DICT[new_node_class][ts_train[-1]]): # 未知ノード同士が仮リンクを張る可能性を省略している。
            if num >= TMP_EDGE_NUM:
                break
            E[new_node, L*num_edge_type*all_node_num + similar_node] = 1
    for row, col in KnownEdge(ts_train[-1]): # LのグラフはL-1のグラフにnew_nodeが加わったもの
        E[row, L*num_edge_type*all_node_num + col] = 1
    return E

def SaveExistTableExtended(ts_train, L, new_node_class_dic):
    # initialize node set
    V = set()
    # add observed node set on train ts
    V |= GetObservedNodes(ts_train[-1], L)
    # add new nodes on test ts
    V |= set(new_node_class_dic.keys())
    # save exist table
    EXIST_TABLE_EXTENDED[list(V), ts_train[0]] = 1
    np.save(OutputDir + '/exist_table', EXIST_TABLE_EXTENDED)
    return len(V), len(GetObservedNodes(ts_train[-1], L)), len(set(new_node_class_dic.keys()))
    
def SaveInputGraphAdjacency(ts_train, ts_test, L, new_node_class_dic):
    # initialize edge set
    E = lil_matrix((all_node_num, (L+1) * num_edge_type * all_node_num))
    # add observed edges set of train ts (L-3, L-2, L-1)
    E = AddObservedEdge(E, ts_train, L)
    ex_nnz = E.nnz
    # add new edges of test ts (L)
    E = AddNewEdge(E, ts_train, ts_test, L, new_node_class_dic)
    new_nnz = E.nnz - ex_nnz
    mmwrite(OutputDir + '/time_series_adjacency/adjacency_' + str(ts_train[0]), E)
    return E.nnz, ex_nnz, new_nnz
```

```python
%%cython
def Matching(list n_list_tup, list t_list_tup, list n, list t):
    cdef list pair_list = []
    cdef set decided_n = set()
    cdef set decided_t = set()
    cdef int idx
    cdef int matched_n, matched_t
            
    while True:
        for idx in range(len(n)):
            matched_n, matched_t = n[idx], t[idx]
            if matched_n in decided_n or matched_t in decided_t:
                continue
            pair_list.append((n_list_tup[matched_n][0], t_list_tup[matched_t][0]))
            decided_n.add(matched_n)
            decided_t.add(matched_t)
            if len(decided_n) == len(n_list_tup):
                break
            if len(decided_t) == len(t_list_tup):
                decided_t = set()
                break
        if len(decided_n) == len(n_list_tup):
            break
    return pair_list
```

```python
def GetNewTeacherVecDic(all_node_attribute, ts_test, L, new_node_class_dic):
    new_vec_dic = {}
    teacher_vec_dic = {}
    known_node_attribute_ts_test = KnownNodeAttribute(ts_test)
    for new_node in new_node_class_dic.keys():
        new_vec_dic[new_node] = all_node_attribute[-1, new_node]
    for teacher_node in GetNodes(ts_test, L, 'new'):
        teacher_vec_dic[teacher_node] = known_node_attribute_ts_test[teacher_node]
    return new_vec_dic, teacher_vec_dic

def BipartiteMatching(new_vec_dic, teacher_vec_dic):
    # sort済みのタプルのリスト[(node_id, vector)]
    n_list_tup = sorted(new_vec_dic.items(), key=lambda x:x[0])
    t_list_tup = sorted(teacher_vec_dic.items(), key=lambda x:x[0])
    
    # 二部グラフ全てのsimilarityを計算
    N = np.array([n_v for n, n_v in n_list_tup])
    T = np.array([t_v for t, t_v in t_list_tup])
    normN = np.sqrt(np.sum(N * N, axis=1))
    normT = np.sqrt(np.sum(T * T, axis=1))
    similarity_matrix = np.dot(N / normN.reshape(-1, 1), (T / normT.reshape(-1, 1)).T)
    
    # similarityのsort
    n, t = np.unravel_index(np.argsort(-similarity_matrix.reshape(-1)), similarity_matrix.shape)
    
    # Greedy Matching
    node_pair_list = Matching(copy.copy(n_list_tup), copy.copy(t_list_tup), n.tolist(), t.tolist())
    
    return node_pair_list
    

def GetNewTeacherDic(node_pair_list):
    new_teacher_dic = {}
    for new, teacher in node_pair_list:
        new_teacher_dic[new] = teacher
    return new_teacher_dic

def GetTeacherNewlsDic(node_pair_list):
    teacher_newls_dic = {}
    for new, teacher in node_pair_list:
        if not teacher in teacher_newls_dic.keys():
            teacher_newls_dic[teacher] = []
        teacher_newls_dic[teacher].append(new)
    return teacher_newls_dic
    
def GetNodePair(all_node_attribute, ts_test, L, new_node_class_dic):
    # Assign teacher nodes to new nodes
    new_vec_dic, teacher_vec_dic = GetNewTeacherVecDic(all_node_attribute, ts_test, L, new_node_class_dic)
    node_pair_list = BipartiteMatching(new_vec_dic, teacher_vec_dic)
    new_teacher_dic = GetNewTeacherDic(node_pair_list)
    teacher_newls_dic = GetTeacherNewlsDic(node_pair_list)
    with open(OutputDir + '/node_pair/new_teacher_dic_' + str(ts_test-L) + '.pickle', mode='wb') as f:
        pickle.dump(new_teacher_dic, f)
    with open(OutputDir + '/node_pair/teacher_newls_dic_' + str(ts_test-L) + '.pickle', mode='wb') as f:
        pickle.dump(teacher_newls_dic, f)
    return new_teacher_dic, teacher_newls_dic

def GetNodePair_(all_node_attribute, ts_test, L, new_node_class_dic): # 未知ノードが出現しないver
    new_teacher_dic = {}
    teacher_newls_dic = {}
    with open(OutputDir + '/node_pair/new_teacher_dic_' + str(ts_test-L) + '.pickle', mode='wb') as f:
        pickle.dump(new_teacher_dic, f)
    with open(OutputDir + '/node_pair/teacher_newls_dic_' + str(ts_test-L) + '.pickle', mode='wb') as f:
        pickle.dump(teacher_newls_dic, f)
    return new_teacher_dic, teacher_newls_dic
```

```python
def SaveLabelEdge(ts_test, ts_train, L, teacher_newls_dic):
    # all_node_numサイズのlabel_edgeを用意
    label_edge = lil_matrix((all_node_num, all_node_num))
    # ts_test時における隣接行列を取得
    known_edge_ts_test = KnownEdge(ts_test)
    # ts_train時に観測されたノード集合を取得
    exist_node_set = GetObservedNodes(ts_train[-1], L)
    
    # ts_testのエッジを4種類に分けているが中身は一緒。
    for row, col in known_edge_ts_test:
        col = col % all_node_num # colはlabel_typeを持つ可能性があるのでall_node_numで割った余りを調べる
        if row in exist_node_set and col in exist_node_set: # exist - exist
            label_edge[row, col] = 1
        elif row in exist_node_set: # exist - new
            if col in teacher_newls_dic.keys(): # teacher node assinged?
                for new in teacher_newls_dic[col]:
                    label_edge[row, new] = 1
        elif col in exist_node_set: # new - exist
            if row in teacher_newls_dic.keys(): # teacher node assinged?
                for new in teacher_newls_dic[row]:
                    label_edge[new, col] = 1
        else: # new - new
            if row in teacher_newls_dic.keys() and col in teacher_newls_dic.keys(): # teacher node assinged?
                for new1 in teacher_newls_dic[row]:
                    for new2 in teacher_newls_dic[col]:
                        label_edge[new1, new2] = 1
    mmwrite(OutputDir + '/label_edge/label_edge_' + str(ts_train[0]), label_edge)

def SaveLabelAttribute(ts_test, ts_train, teacher_newls_dic):    
    label_attribute = np.zeros((all_node_num, attribute_size))
    known_node_attribute_ts_test = KnownNodeAttribute(ts_test)
    label_attribute[:known_node_num] = known_node_attribute_ts_test
    
    # teacher idxはゼロベクトルで上書きし、new idxに移動させる。
    label_attribute[list(GetNodes(ts_test, L, 'new'))] = np.zeros(attribute_size)
    for teacher_node, newls in teacher_newls_dic.items():
        for new_node in newls:
            label_attribute[new_node] = known_node_attribute_ts_test[teacher_node]
    np.save(OutputDir + '/label_node_attribute/label_node_attribute_' + str(ts_train[0]), label_attribute)

def SaveLabelLost(ts_test, L):
    label_lost = np.zeros(known_node_num)
    for node in GetNodes(ts_test, L, 'lost'):
        label_lost[node] = 1
    np.save(OutputDir + '/label_node/label_lost_' + str(ts_test-L), label_lost)

def SaveLabelReturn(ts_test, L):
    label_return = np.zeros(known_node_num)
    for node in GetNodes(ts_test, L, 'return'):
        label_return[node] = 1
    np.save(OutputDir + '/label_node/label_return_' + str(ts_test-L), label_return)
```
